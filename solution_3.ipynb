```python
# Function to load all sheets from a file
def load_data(file_path, year):
    sheets = pd.read_excel(file_path, sheet_name=None)  # Load all sheets into a dictionary
    data_list = []
    for school_name, df in sheets.items():  # Iterate over each sheet
        df['Year'] = year  # Add a Year column
        df['School'] = school_name  # Add a School column
        data_list.append(df)  # Add processed DataFrame to the list
    return pd.concat(data_list, ignore_index=True)  # Combine all sheets into a single DataFrame
```

### 1. Reward the top performer (student) of each school based on cumulative marks scored in the last three years for all subjects

### 3.Find out students with the highest improvement for each subject from 2019-21 combining all the schools together

```python

marks_columns = ['Hindi', 'English', 'Mathematics', 'Physics', 'Chemistry', 
                 'Biology', 'History', 'Geography', 'Civics', 
                 'Computer Science', 'Physical Education']

# Step 2: Create a DataFrame for improvements
improvement_data = data_all.pivot_table(
    index=['Student Name'],  # Group by students
    columns='Year',
    values=marks_columns,
    aggfunc='sum'
)

# Flatten columns
improvement_data.columns = ['_'.join(map(str, col)) for col in improvement_data.columns]

# Ensure 2019, 2020, and 2021 data exist, fill missing values with 0
for year in [2019, 2020, 2021]:
    for subject in marks_columns:
        col_name = f"{subject}_{year}"
        if col_name not in improvement_data.columns:
            improvement_data[col_name] = 0

# Calculate improvements
improvement_records = []
for subject in marks_columns:
    # Improvements for 2019  2020
    improvement_data[f"{subject}_Improvement_2019_2020"] = (
        improvement_data[f"{subject}_2020"] - improvement_data[f"{subject}_2019"]
    )
    # Improvements for 2020  2021
    improvement_data[f"{subject}_Improvement_2020_2021"] = (
        improvement_data[f"{subject}_2021"] - improvement_data[f"{subject}_2020"]
    )
    
    # Step 3: Find the student with the highest improvement for each subject
    # 2019  2020
    max_2019_2020 = improvement_data[f"{subject}_Improvement_2019_2020"].idxmax()
    max_2019_2020_val = improvement_data[f"{subject}_Improvement_2019_2020"].max()
    improvement_records.append({
        "Subject": subject,
        "Year Transition": "2019  2020",
        "Student Name": max_2019_2020,
        "Improvement": max_2019_2020_val
    })
    
    # 2020  2021
    max_2020_2021 = improvement_data[f"{subject}_Improvement_2020_2021"].idxmax()
    max_2020_2021_val = improvement_data[f"{subject}_Improvement_2020_2021"].max()
    improvement_records.append({
        "Subject": subject,
        "Year Transition": "2020  2021",
        "Student Name": max_2020_2021,
        "Improvement": max_2020_2021_val
    })

# Step 4: Convert the improvement records into a DataFrame
highest_improvements = pd.DataFrame(improvement_records)

# Display the DataFrame
print(highest_improvements)
```

### 4.Identify best school for Arts, Science and Commerce streams based on marks scored by students in respective subjects for those streams in last three years

```python
# Stream-wise subject lists
streams = {
    "Arts": [ 'Hindi', 'English', 'Physical Education'],
    "Science": ['Mathematics', 'Physics', 'Chemistry', 'Biology'],
    "Commerce": ['History', 'Geography', 'Civics', 'Computer Science']
}
```

```python

# Calculate total marks per stream for each school
stream_scores = {}
for stream, subjects in streams.items():
    stream_scores[stream] = data_all.groupby(['School', 'Year'])[subjects].sum().mean(axis=1).reset_index()
    stream_scores[stream].columns = ['School', 'Year', 'Average Marks']
    
print("Best Schools by Stream:", stream_scores)
# Identify the best school for each stream across all years
best_schools = {}
for stream, df in stream_scores.items():
    best_school = df.groupby('School')['Average Marks'].mean().idxmax()
    best_schools[stream] = best_school

print("Best Schools by Stream:", best_schools)
```

```python

best_schools_stream = pd.DataFrame({
    'Stream': ['Arts', 'Science', 'Commerce'], 
    'Best School': ['Birla HS', 'International', 'Birla HS']
})

# Use a barplot to show counts of the best school for each stream
plt.figure(figsize=(8, 6))
sns.barplot(
    x='Stream',
    y=[1, 1, 1],  # Assign dummy counts to display the categorical relationship
    hue='Best School',
    data=best_schools_stream,
    dodge=False,
    palette='tab10'
)
plt.title('Best School for Each Stream')
plt.xlabel('Stream')
plt.ylabel('Count (1 per Stream)')
plt.legend(title='Best School')
plt.tight_layout()
plt.show()
```

### 5. Categorize students marks into 5 sections and calculate counts per school.

```python
# Define categories
bins = [0, 20, 40, 60, 80, 100]
labels = ['Very Poor', 'Poor', 'Average', 'Good', 'Very Good']

# Assign categories to each subject's average marks
data_all['Category'] = data_all[marks_columns].mean(axis=1).apply(lambda x: pd.cut([x], bins=bins, labels=labels)[0])

# Count categories for each school per year
category_counts = data_all.groupby(['School', 'Year', 'Category']).size().reset_index(name='Count')
print(category_counts)
```

```python
# Stacked bar chart for categories
category_counts_pivot = category_counts.pivot_table(index='School', columns='Category', values='Count', aggfunc='sum', fill_value=0)

category_counts_pivot.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20c')
plt.title('Student Categories Per School (Average Marks)', fontsize=16)
plt.ylabel('Number of Students', fontsize=12)
plt.xlabel('School', fontsize=12)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 6.Which is the best school for each year 2019, 2020 and 2021 based on highest no. of students in Good and Very Good category

```python
# Filter students in Good and Very Good categories
good_students = category_counts[category_counts['Category'].isin(['Good', 'Very Good'])]

# Find the school with the most students in these categories for each year
best_schools_yearly = good_students.groupby(['Year', 'School'])['Count'].sum().reset_index()
best_schools = best_schools_yearly.loc[best_schools_yearly.groupby('Year')['Count'].idxmax()]
print(best_schools)
```

```python
# Line plot for Good and Very Good categories
good_counts_yearly = good_students.groupby(['Year', 'School'])['Count'].sum().unstack()

good_counts_yearly.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Good and Very Good Categories Per Year (Best School)', fontsize=16)
plt.ylabel('Number of Students', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.legend(title='School', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 7.Fastest-growing school (Overall and Streamwise)

```python
# Calculate total marks for each school across years
growth_data = data_all.groupby(['School', 'Year'])[marks_columns].sum().sum(axis=1).unstack()

# Calculate growth rate for each school
growth_data['Growth Rate'] = (growth_data[2021] - growth_data[2019]) / growth_data[2019]

# Identify the fastest-growing school overall
fastest_school = growth_data['Growth Rate'].idxmax()

# For streams
stream_growth = {}
for stream, subjects in streams.items():
    stream_scores = data_all.groupby(['School', 'Year'])[subjects].mean().sum(axis=1).unstack()
    stream_scores['Growth Rate'] = (stream_scores[2021] - stream_scores[2019]) / stream_scores[2019]
    stream_growth[stream] = stream_scores['Growth Rate'].idxmax()
print(f"Fastest Growing School (Overall): {fastest_school}")
print(f"Streamwise Growth: {stream_growth}")
```

```python
# Bar chart for fastest-growing school
growth_data_sorted = growth_data.sort_values(by='Growth Rate', ascending=False).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=growth_data_sorted, x='School', y='Growth Rate', palette='viridis')
plt.title('Fastest Growing Schools (Overall Growth Rate)', fontsize=16)
plt.xticks(rotation=45)
plt.ylabel('Growth Rate (%)', fontsize=12)
plt.xlabel('School', fontsize=12)
plt.tight_layout()
plt.show()

# Streamwise growth visualization
stream_growth_df = pd.DataFrame.from_dict(stream_growth, orient='index', columns=['School'])
stream_growth_df.reset_index(inplace=True)
stream_growth_df.columns = ['Stream', 'Fastest Growing School']
print(stream_growth_df)
```
